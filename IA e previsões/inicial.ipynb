### Case: Score de Crédito dos Clientes

Você foi contratado por um banco para conseguir definir o score de crédito dos clientes. Você precisa analisar todos os clientes do banco e, com base nessa análise, criar um modelo que consiga ler as informações do cliente e dizer automaticamente o score de crédito dele: Ruim, Ok, Bom

Arquivos da aula: https://drive.google.com/drive/folders/1FbDqVq4XLvU85VBlVIMJ73p9oOu6u2-J?usp=drive_link

```python
import pandas as pd

tabela = pd.read_csv("clientes.csv")

display(tabela)

display(tabela.info())
```

```python
from sklearn.preprocessing import LabelEncoder
#tem como atribuir uma função a uma variavel
codificador = LabelEncoder()

#Ele transforma todas as palavra em numero ou seja ele pega cada tipo de dado que tem na coluna profissão
#E transformar em numero,LabelEncoder().fit_transform é a função dai so passar as colunas
tabela["profissao"] = codificador.fit_transform(tabela["profissao"])
tabela["mix_credito"] = codificador.fit_transform(tabela["mix_credito"])
tabela["comportamento_pagamento"] = codificador.fit_transform(tabela["comportamento_pagamento"])

display(tabela)
```

```python
#Exclui e separa as colunas que vão ser usados na IA
y = tabela["score_credito"]
colunas = ["score_credito", "id_cliente"]

x = tabela.drop(columns=colunas)

#metodo que divide o treiono e o teste
from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

#cria a IA
modelo_arvoreDecisao = RandomForestClassifier()
modelo_knn = KNeighborsClassifier()

#passa os dados de treino para a IA treinar
modelo_arvoreDecisao.fit(x_treino, y_treino)
modelo_knn.fit(x_treino, y_treino)
```

```python
from sklearn.metrics import accuracy_score

#Faz o teste da IA com os dados teste
previsao_arvoreDecesao = modelo_arvoreDecisao.predict(x_teste)
previsao_knn = modelo_knn.predict(x_teste.to_numpy())

#compara a previsão do banco com a da IA
print(accuracy_score(y_teste, previsao_arvoreDecesao))
print(accuracy_score(y_teste, previsao_knn))
```

```python
novos_clientes = pd.read_csv("novos_clientes.csv")
display(novos_clientes)
novos_clientes["profissao"] = codificador.fit_transform(novos_clientes["profissao"])
novos_clientes["mix_credito"] = codificador.fit_transform(novos_clientes["mix_credito"])
novos_clientes["comportamento_pagamento"] = codificador.fit_transform(novos_clientes["comportamento_pagamento"])

previsao = modelo_arvoreDecisao.predict(novos_clientes)
display(previsao)
```
